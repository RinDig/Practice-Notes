9. (6 points)  Not only are you taking CIS 340 this term, but you are also taking Calculus 1. The syllabus in that class indicates that your grade is computed as follows:

| Assignment | Percentage of Total Grade |
| :------------- | :----------: |
| Exam 1 | 15% |

| Homework | 25% |
| Big Project | 15% |
| Final Exam | 25% |

(a) You received an 82 on Exam 1, a 91 on Exam 2, a Homework grade of 75, an 87 on the Big Project, and an 84 on the final. Assign variable names that make sense to all of these values and use the table above to compute your overall grade using the variables.

```python
Exam_1=.82*.15
Exam_2=.91*.20
Homework=.75*.25
Big_Project=.87*.15
Final_Exam=.84*.25
Overall_grade=Exam_1+Exam_2+Homework+Big_Project+Final_Exam
print(Overall_grade*100,'%')
```
